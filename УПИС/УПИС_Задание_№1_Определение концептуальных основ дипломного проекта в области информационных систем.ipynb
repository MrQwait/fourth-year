# **Задание 1. Определение концептуальных основ дипломного проекта в области информационных систем**



В рамках дисциплины "Управление проектами информационных систем" представьте развернутое концептуальное обоснование вашего проекта, ответив на следующие вопросы:


#### **1. Обоснование необходимости разработки**


Для кого предназначено разрабатываемое Вами решение и какие потребности целевой аудитории оно призвано удовлетворить?




*Обоснуйте актуальность решаемой проблемы для конкретных заинтересованных сторон.*

Решение предназначено для игроков, которым требуется уникальный и непредсказуемый игровой опыт при каждом запуске, а также для разработчиков инди-игр, стремящихся автоматизировать создание многоуровневых локаций.
С моей стороны проект представляет собой исследовательскую работу по адаптации и практической реализации алгоритмов процедурной генерации в трёхмерном пространстве, с целью решения задачи вертикальной связности подземелий без ручного проектирования.
Результаты работы могут быть использованы как основа для дальнейших исследований в области генеративного дизайна игр.

---


#### **2. Ожидаемые результаты проекта**



Опишите конкретные результаты, которые планируется получить по завершении проекта. Какой программный продукт, методика или информационная система будет создан/создана в результате вашей работы?

Будет создан прототип игры на Unity с системой процедурной генерации трёхмерных подземелий. Система использует триангуляцию Делоне для определения связей между комнатами, строит минимальное остовное дерево для гарантии проходимости, добавляет случайные коридоры для увеличения сложности, а затем прокладывает пути между комнатами с помощью модифицированного алгоритма поиска.
Результат — рабочая сцена с автоматически генерируемыми комнатами, коридорами и вертикальными соединениями между этажами, готовая к интеграции с механиками исследования и сбора предметов.

---

#### **3. Методология использования**


Каким образом предполагается практическое применение разрабатываемого решения? Опишите основные сценарии взаимодействия пользователей с разрабатываемым решением.

Генератор работает как фоновый модуль при старте игры. Он последовательно размещает комнаты в 3D-сетке, добавляет дополнительные коридоры для вариативности, а затем прокладывает соединения между комнатами с учётом ограничений движения между этажами.
Пользователь не взаимодействует с генерацией — он сразу попадает в сгенерированный уровень, где должен исследовать пространство, находить ключи, избегать ловушек и собирать сокровища.

---

#### **4. Входные данные и их источники**


Какие типы данных будут использоваться в качестве входных данных для вашего решения? В каком формате они будут представлены и из каких источников планируется их получение?

Входными данными являются параметры генерации: размеры трёхмерной сетки (например, 30×5×30), количество комнат, вероятность добавления дополнительных коридоров, а также соотношение подъёма и длины лестниц. Все параметры задаются в конфигурационных файлах или скриптах Unity при запуске проекта. Источник данных — ручная настройка разработчиком, без внешних источников.

---

#### **5. Выходные данные и практическая значимость**


Какие результаты будет формировать Ваше решение на выходе?

На выходе формируется полностью функциональная 3D-сцена с комнатами, коридорами и соединениями между этажами, в которой реализованы основы игрового процесса: перемещение, обнаружение предметов, взаимодействие с объектами.

---

Опишите конкретно и без "воды" пользу (практическую ценность), которую заинтересованные стороны получат от использования этих результатов.

Система позволяет быстро создавать уникальные уровни для игр жанра данжин-кроулер без ручного дизайна, сохраняя баланс между проходимостью и сложностью. Это снижает затраты на разработку контента, повышает повторяемость игры и демонстрирует применение вычислительной геометрии (триангуляция Делоне) для решения реальных задач геймдева.
