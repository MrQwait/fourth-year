## 1. ИНФОРМАЦИОННАЯ АРХИТЕКТУРА

### 1.1. Карта навигации (структура системы)

```python
Главное меню
├── Продолжить ──→ Игровой экран [ИГРА]
│               ├── Инвентарь
│               └── Меню паузы
│                   ├── Продолжить
│                   ├── Настройки
│                   └── Выход в главное меню
├── Новая игра ──→ Настройка генерации ──→ Меню с покупками и улучшениями ──→ Запуск генерации
│                                         ├── Улучшения Осветителя           │
│                                         ├── Покупка оружия                 │
│                                         ├── Покупка расходников            │
│                                         └── Установка модификаторов        │
│                                                                            └── ──→ Игровой экран [ИГРА]
│                                                                            ├── Инвентарь
│                                                                            └── Меню паузы
│                                                                                ├── Продолжить
│                                                                                ├── Настройки
│                                                                                └── Выход в главное меню
├── Настройки
└── Выйти из игры
```

### 1.2. Пользовательские пути


**Сценарий 1: Полное прохождение уровня**  
1. Запуск игры → **Главное меню**.  
2. Выбор **Новая игра** → **Настройка генерации**.  
3. Переход в **Меню с покупками и улучшениями** → выбор **Улучшения Осветителя**, **Покупка оружия**, **Покупка расходников**, **Установка модификаторов**.  
4. Запуск генерации → **Игровой экран [ИГРА]**
5. Прохождение уровня
6. Завершение уровня → **Главное меню**.

**Сценарий 2: Продолжение сохранённой сессии**  
1. Запуск игры → **Главное меню**.  
2. Выбор **Продолжить** → переход в **Игровой экран [ИГРА]**.  
3. Продолжение прохождения уровня  
4. Завершение уровня → **Главное меню**

**Сценарий 3: Настройка и подготовка перед запуском**  
1. **Главное меню** → **Новая игра** → **Настройка генерации** → изменение параметров
2. Переход в **Меню с покупками и улучшениями** → покупка/установка модулей → тестирование баланса.  
3. Запуск **Игровой экран [ИГРА]** → проверка генерации и механик.  
4. Пауза (Esc) → **Меню паузы** → **Настройки** → корректировка управления → возврат в игру или выход.  
5. Завершение → **Главное меню** → повтор цикла для итераций.

## 6. ОБОСНОВАНИЕ ПРОЕКТНЫХ РЕШЕНИЙ

### 6.1. Соответствие принципам юзабилити

| Принцип                          | Реализация в интерфейсе                                                                      | Пример / Конкретный элемент                              |
|----------------------------------|----------------------------------------------------------------------------------------------|----------------------------------------------------------|
| Видимость состояния              | Все ключевые параметры игрока всегда видны на экране, без необходимости открывать меню      | HUD в левом нижнем углу: здоровье, щит, заряд Осветителя, таймер уровня |  |
| Свобода действий пользователя    | Игрок в любой момент может выйти без потери прогресса, отменить покупки до запуска уровня    | Кнопка «Выход в главное меню» в Меню паузы, отмена в магазине |
| Минимизация нагрузки на память   | Все действия доступны через 1–2 клика/клавиши, нет скрытых комбинаций                       | Esc → Меню паузы, I → Инвентарь                          |
| Предотвращение ошибок            | Подтверждение при дорогих покупках и сбросе прогресса                                        | Диалог «Вы уверены?» при покупке модуля 3-го уровня     |

### 6.2. Применение законов проектирования
**Закон Фиттса**  
Часто используемые элементы сделаны крупными и расположены в углах/краях экрана, куда курсор или взгляд попадает быстрее:
- HUD-индикаторы закреплены в левом нижнем углу
- Кнопка «Продолжить» в Меню паузы — самая крупная и вверху списка

**Закон Хика**  
Количество выборов сведено к минимуму:
- В главном меню — всего 4 кнопки
- В Меню паузы — 3 действия
- В магазине — 4 вкладки вместо 10–15 отдельных кнопок

**Закон Миллера**  
Информация сгруппирована по 3–6 элементов:
- HUD содержит ровно 6 индикаторов (здоровье, щит, заряд, искажения, оружие, время)
- Вкладки магазина: 4 категории (Улучшения Осветителя, Оружие, Расходники, Модификаторы)


### 6.3. Улучшения относительно аналогов
##### **Пример 1**  
**Проблема в аналогах:** Warframe и Deep Rock Galactic перегружают экран десятками индикаторов и цифр, из-за чего новичку сложно понять своё состояние.  
**Наше решение:** Минималистичный HUD — только 5 критически важных индикаторов без лишних цифр и иконок, всё читаемо даже на 720p.

##### **Пример 2**  
**Проблема в аналогах:** В Deep Rock Galactic и Immortal Redneck нет постоянного таймера уровня и индикатора прогресса.  
**Наше решение:** Таймер уровня и счётчик собранных искажений всегда видны в HUD, игрок сразу понимает, сколько осталось.

##### **Пример 3**  
**Проблема в аналогах:** В Immortal Redneck и Vaporum нельзя отменить покупку или изменить настройки после подтверждения.  
**Наше решение:** Кнопка «Отмена» во всех экранах магазина и настроек генерации, изменения применяются только при запуске уровня.

##### **Пример 4**  
**Проблема в аналогах:** На высоких разрешениях элементы управления у Warframe становятся слишком мелкими или растянутыми.  
**Наше решение:** Canvas Scaler + якоря гарантируют читаемость и кликабельность от 1280×720 до 2560×1440.

## 7. СПЕЦИФИКАЦИЯ ИНТЕРФЕЙСА

### 7.1. Сетка и отступы
- **Базовая сетка:** 8 px
- **Колонки:** 12-колоночная система
- **Отступы:**
  - По краям экрана: **24 px**
  - Между элементами внутри блоков: **16 px**
  - Внутренние отступы кнопок и полей: **16 px** горизонтально, **12 px** вертикально

### 7.2. Компоненты интерфейса

**Кнопки**
- **Размер:**
  - Основные (главное меню, «Продолжить»): **320 × 64 px**
  - Второстепенные («Отмена», «Назад»): **160 × 56 px**
  - Вкладки в магазине: **180 × 48 px**
- **Скругление:** 12 px
- **Состояния:**
  - Обычное: текст `#FFFFFF`
  - Наведение: текст `#ededed`

### 7.3. Анимации и переходы

- **Длительность:**
  - Появление/исчезновение меню: **250 ms**
  - Наведение/нажатие кнопок: **150 ms**

- **Функция плавности:**
  - Появление: **ease-out-quart**
  - Исчезновение: **ease-in-quart**

- **Типы анимаций:**
  - **Появление/исчезновение меню:** Canvas Group Alpha 0 → 1 + Scale 0.95 → 1.0
  - **Открытие Меню паузы:** затемнение фона (0 → 0.7 чёрный) + выдвижение панели сверху вниз
  - **Кнопки:** наведение — свечение + подъём (Y +4 px); нажатие — вдавливание (scale 0.95)
  - **Инвентарь:** slide-in справа (X от +400 px → 0)
  - **Переходы между экранами:** чёрный оверлей 0 → 1 → 0
