### **3. Диаграмма компонентов (Component Diagram)**

Диаграмма компонентов.PNG

**Ядро игры**  
Управляет общим жизненным циклом игры: запуск, пауза, завершение
**Генератор уровней**    
Отвечает за процедурную генерацию подземелий на основе заданных параметров (количество комнат, сложность). Размещает врагов, ловушки и загадки.  
**Система игрока**    
Управление персонажем, ведение инвентаря и боевая механика. Обеспечивает взаимодействие игрока с игровым миром.  
**Модуль врагов и ИИ**  
Поведение врагов (спавн, патрулирование, преследование и атаку игрока)  
**Интерактивные объекты**
Логика ловушек, загадок и других объектов, с которыми может взаимодействовать игрок.  
**Интерфейс и магазин**  
Пользовательский интерфейс (главное меню, HUD и систему покупки расходников до и во время игры)  
**Данные и настройки**  
Настройки игры (сложность, количество комнат), а также сохранение и загрузка игрового прогресса.
